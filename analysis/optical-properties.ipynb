## Plot optical properties

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(7, 3.5))
# ----------------------------------#
# plot refractive index
# ----------------------------------#
axes[0].plot(EnGaAs, RefractiveGaAs, color='blue', ls='-', lw=1.5, label='GaAs')
axes[0].plot(EnGaSb, RefractiveGaSb, color='green', ls='-', lw=1.5, label='GaSb')
axes[0].plot(EnInAs, RefractiveInAs, color='red', ls='-', lw=1.5, label='InAs')
axes[0].plot(EnInP, RefractiveInP, color='orange', ls='-', lw=1.5, label='InP')
# X-axis
axes[0].set_xlim(0.0, 15.0)
axes[0].set_xlabel('$\hbar \omega$ (eV)')
# Y-axis
axes[0].set_ylim(0.0, 5.5)
axes[0].set_ylabel('$n(\hbar \omega)$')
axes[0].legend(loc='best', prop={'size': 10})
# ----------------------------------#
# plot extinction coefficient
# ----------------------------------#
axes[1].plot(EnGaAs, ExtinctionGaAs, color='blue', ls='-', lw=1.5, label='GaAs')
axes[1].plot(EnGaSb, ExtinctionGaSb, color='green', ls='-', lw=1.5, label='GaSb')
axes[1].plot(EnInAs, ExtinctionInAs, color='red', ls='-', lw=1.5,label='InAs')
axes[1].plot(EnInP, ExtinctionInP, color='orange', ls='-', lw=1.5, label='InP')
# X-axis
axes[1].set_xlim(0.0, 15.0)
axes[1].set_xlabel('$\hbar \omega$ (eV)')
# Y-axis
axes[1].set_ylim(0.0, 5.5)
axes[1].set_ylabel('$\kappa(\hbar \omega)$')
axes[1].legend(loc='best', prop={'size': 10})

plt.tight_layout()

# save figure
fig.savefig('./imag/optical.svg')

plt.show()
```

```python
# ----------------------------------#
# plot absorption coefficient
# ----------------------------------#
# plotting
fig = plt.figure(figsize=(4.5, 3.5))

plt.plot(EnGaAs, AbsorptionGaAs*1.e-8, color='blue', ls='-', lw=1.5, label='GaAs')
plt.plot(EnGaSb, AbsorptionGaSb*1.e-8, color='green', ls='-', lw=1.5, label='GaSb')
plt.plot(EnInAs, AbsorptionInAs*1.e-8, color='red', ls='-', lw=1.5, label='InAs')
plt.plot(EnInP, AbsorptionInP*1.e-8, color='orange', ls='-', lw=1.5, label='InP')

# X-axis
plt.xlabel('$\hbar \omega$ (eV)')
plt.xlim(0.0, 15.0)

# Y-axis
plt.ylabel(r'$\alpha \, (10^{8}\, \mathrm{m}^{-1})$')
plt.ylim(0.0, 2.1)

plt.legend(loc='best', prop={'size': 10})

plt.tight_layout()

# save figure
fig.savefig('./imag/absorption.svg')

plt.show()
```
