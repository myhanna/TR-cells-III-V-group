```python
# numerical module
import numpy as np

#setup plotting 
import matplotlib.pyplot as plt
plt.style.use('SciencePlots-APS.mplstyle')
```

```python
# Data
Material = ['GaAs', 'GaSb', 'InAs', 'InP']
EgapMaterial = np.array([1.016, 0.554, 0.209, 1.240])

# efficiency data summary
## Realistic model
OurWork_Ideal = np.array([66.75388803313726, 63.62587203882166, 55.53044292124745, 67.38136308408829])
OurWork_allLosses = np.array([0.18817028394955046, 4.583054705665605, 24.98972179660062, 0.03181364966136108])
## DBM
DBM_Ideal = np.array([67.45574902555636, 65.11438988304258, 56.89707798113505, 67.94305849718299])
DBM_allLosses = np.array([0.006959615155465987, 0.5394896894752157, 11.58566085431847, 0.0007463970839776238])
```

```python
# Plotting
x = np.arange(len(Material))  # X-axis locations for bars
width = 0.35  # Width of the bars

# Bar chart for Eta with out losses
fig, ax = plt.subplots(nrows=2, ncols=1, figsize=(7, 7))
rects1 = ax[0].bar(x - width/2, OurWork_Ideal, width, label='This work')
rects2 = ax[0].bar(x + width/2, DBM_Ideal, width, label='Detailed balance model')

# Add text annotations for each bar
for i, (r1, r2) in enumerate(zip(rects1, rects2)):
    ax[0].text(r1.get_x() + r1.get_width() / 2, r1.get_height(), f"{OurWork_Ideal[i]:.2f}", ha='center', va='bottom',fontsize=10, color='blue')
    ax[0].text(r2.get_x() + r2.get_width() / 2, r2.get_height(), f"{DBM_Ideal[i]:.2f}", ha='center', va='bottom', fontsize=10, color='green')

# Add labels and title
ax[0].set_ylabel('Efficiency w/o loss (%)', fontsize=16)
ax[0].set_xticks(x)
ax[0].set_xticklabels(Material, fontsize=16)
ax[0].set_ylim(0.0, 75.0)
ax[0].set_title(r'$T_{cell}$ = 1000 K')


# Bar chart for EtaWLLoss
rects1 = ax[1].bar(x - width/2, OurWork_allLosses, width, label='This work')
rects2 = ax[1].bar(x + width/2, DBM_allLosses, width, label='Detailed balance model')

# Add text annotations for each bar
for i, (r1, r2) in enumerate(zip(rects1, rects2)):
    ax[1].text(r1.get_x() + r1.get_width() / 2, r1.get_height(), f"{OurWork_allLosses[i]:.2f}", ha='center', va='bottom', fontsize=10, color='blue')
    ax[1].text(r2.get_x() + r2.get_width() / 2, r2.get_height(), f"{DBM_allLosses[i]:.2f}", ha='center', va='bottom', fontsize=10, color='green')

# Add labels, title, and legend
ax[1].set_ylabel('Efficiency w. loss (%)', fontsize=16)
ax[1].set_xticks(x)
ax[1].set_xticklabels(Material, fontsize=16)
ax[1].set_ylim(0.0, 35.0)

# Move the legend outside the figure
ax[1].legend(loc='upper center', bbox_to_anchor=(0.5, -0.15), fancybox=True, shadow=True, ncol=2)

fig.tight_layout()

fig.savefig('./imag-rev/comparison-TRcell-performance.svg')

plt.show()
```
